Proyek ini mengintegrasikan semua materi Python:

* **Perulangan** (memproses data)

* **Visualisasi Data** (Matplotlib)

**Tujuan:** Menghitung nilai akhir (bobot 30/30/40), menentukan kelulusan (syarat >=60), menyaring mahasiswa terbaik, dan menampilkan grafik distribusi nilai.

## Import Library

```python
# 3. Import Library
import pandas as pd # Untuk handling CSV
import matplotlib.pyplot as plt # Untuk visualisasi
```

## Load Data CSV

```python
# 4. Load Data & Exception Handling (FileNotFoundError)
df = None
file_path = 'nilai_mahasiswa.csv'

try:
    df = pd.read_csv(file_path)
    print(f"Data berhasil dimuat dari {file_path}. Jumlah baris: {len(df)}")
    print("\n5 Baris Data Pertama:")
    print(df.head())
except FileNotFoundError:
    print(f"ERROR: File '{file_path}' tidak ditemukan.")
    print("Pastikan Anda sudah membuat file 'nilai_mahasiswa.csv' di folder yang sama.")
except Exception as e:
    print(f"ERROR saat memuat data: Terjadi kesalahan lain ({e})")
```

## Preprocessing

```python
# 5. Preprocessing (Opsional: Bersihkan data dan pastikan tipe data benar)
if df is not None:
    # Hapus baris dengan nilai kosong (NaN)
    df = df.dropna()
    # Konversi kolom nilai menjadi tipe numerik
    try:
        df['Nilai_Tugas'] = pd.to_numeric(df['Nilai_Tugas'], errors='coerce')
        df['Nilai_UTS'] = pd.to_numeric(df['Nilai_UTS'], errors='coerce')
        df['Nilai_UAS'] = pd.to_numeric(df['Nilai_UAS'], errors='coerce')
        
        # Hapus baris yang gagal dikonversi (menjadi NaN lagi)
        df = df.dropna()
        print("\nPreprocessing Selesai: Nilai diubah ke tipe numerik dan baris kosong/invalid dihapus.")
    except Exception as e:
        print(f"ERROR Preprocessing: Gagal mengkonversi nilai ke numerik. ({e})")

# Cek kembali jika dataframe masih valid setelah preprocessing
if df.empty:
    print("WARNING: Dataframe kosong setelah preprocessing.")
```

## Fungsi dan OOP

```python
# 6. Fungsi (def & lambda) dan OOP (Kelas)

# OOP: Kelas Mahasiswa
class Mahasiswa:
    def __init__(self, nama, tugas, uts, uas):
        self.nama = nama
        self.nilai_tugas = tugas
        self.nilai_uts = uts
        self.nilai_uas = uas
        self.nilai_akhir = self.hitung_nilai_akhir()
        self.status = self.cek_kelulusan()

    # Fungsi def: Menghitung nilai akhir dengan bobot
    def hitung_nilai_akhir(self):
        try:
            # Bobot: Tugas 30%, UTS 30%, UAS 40%
            akhir = (self.nilai_tugas * 0.3) + (self.nilai_uts * 0.3) + (self.nilai_uas * 0.4)
            # Menggunakan percabangan sederhana di dalam fungsi
            return round(akhir, 2)
        except TypeError:
            # Menangani jika ada data non-numerik yang entah bagaimana lolos
            return 0.0

    # Percabangan di dalam metode kelas: Menentukan kelulusan
    def cek_kelulusan(self):
        if self.nilai_akhir >= 60:
            return "Lulus"
        else:
            return "Tidak Lulus"

# Fungsi lambda: Untuk filter mahasiswa terbaik (Nilai Akhir > 80)
filter_terbaik = lambda mhs: mhs.nilai_akhir > 80

print("Kelas Mahasiswa dan Fungsi Kalkulasi Siap.")
```

## Percabangan & Perulangan

```python
# 7. Percabangan & Perulangan
mahasiswa_list = []

if df is not None and not df.empty:
    # Perulangan: Memproses setiap baris data
    for index, row in df.iterrows():
        try:
            # Percabangan (disini: pengecekan nilai harus positif)
            if row['Nilai_Tugas'] < 0 or row['Nilai_UTS'] < 0 or row['Nilai_UAS'] < 0:
                print(f"Data invalid (nilai negatif) dilewati untuk: {row['Nama']}")
                continue
            
            # Membuat objek OOP Mahasiswa
            mhs = Mahasiswa(
                nama=row['Nama'], 
                tugas=row['Nilai_Tugas'], 
                uts=row['Nilai_UTS'], 
                uas=row['Nilai_UAS']
            )
            mahasiswa_list.append(mhs)
            
        # Exception Handling (jika ada data invalid yang lolos dari preprocessing)
        except ValueError:
            print(f"Data invalid (bukan angka) dilewati untuk: {row['Nama']}")
        except KeyError as ke:
            print(f"ERROR: Kolom tidak ditemukan: {ke}. Pastikan kolom CSV sesuai.")
            break
        except Exception as e:
            print(f"ERROR tak terduga saat memproses {row['Nama']}: {e}")

    # Mengubah list objek OOP kembali menjadi DataFrame
    data_hasil = [
        {
            'Nama': mhs.nama, 
            'Nilai_Akhir': mhs.nilai_akhir, 
            'Status': mhs.status
        } for mhs in mahasiswa_list
    ]
    df_hasil = pd.DataFrame(data_hasil)

    # Fungsi Lambda: Menerapkan filter
    mahasiswa_terbaik = list(filter(filter_terbaik, mahasiswa_list))

    print("\n--- Hasil Analisis ---")
    print("Data hasil kalkulasi (Nilai Akhir & Status Kelulusan):")
    print(df_hasil)
    
    print("\nMahasiswa Terbaik (Nilai Akhir > 80):")
    if mahasiswa_terbaik:
        for mhs in mahasiswa_terbaik:
            print(f"- {mhs.nama} (Nilai Akhir: {mhs.nilai_akhir})")
    else:
        print("- Tidak ada mahasiswa dengan Nilai Akhir di atas 80.")
else:
    df_hasil = pd.DataFrame()
    print("Tidak ada data untuk diproses. Periksa file CSV Anda.")
```

## Exception Handling

```python
# 8. Exception Handling (Menyimpan File)
if not df_hasil.empty:
    output_path = 'hasil_analisis.csv'
    try:
        # File Handling: Menulis hasil ke CSV baru
        df_hasil.to_csv(output_path, index=False)
        print(f"\nBerhasil menyimpan hasil analisis ke '{output_path}'")
    except Exception as e:
        print(f"FATAL ERROR: Gagal menyimpan file hasil. ({e})")
```

## Visualisasi Data

```python
# 9. Visualisasi Data
if not df_hasil.empty:
    
    # Visualisasi 1: Grafik batang distribusi nilai akhir
    print("\n--- Grafik Distribusi Nilai Akhir ---")
    
    # Hitung rata-rata
    rata_rata = df_hasil['Nilai_Akhir'].mean()
    
    plt.figure(figsize=(14, 7)) # Ukuran figure lebih besar agar nama tidak bertumpuk
    plt.bar(df_hasil['Nama'], df_hasil['Nilai_Akhir'], color='skyblue')
    plt.axhline(60, color='red', linestyle='--', linewidth=1, label='Batas Lulus (60)')
    plt.axhline(rata_rata, color='green', linestyle='-', linewidth=2, label=f'Rata-rata ({rata_rata:.2f})')
    
    plt.xlabel('Nama Mahasiswa')
    plt.ylabel('Nilai Akhir')
    plt.title('Grafik Batang Nilai Akhir Mahasiswa')
    plt.xticks(rotation=45, ha='right') # Memutar label nama agar mudah dibaca
    plt.legend()
    plt.grid(axis='y', linestyle=':')
    plt.tight_layout()
    plt.show()

    # Visualisasi 2: Pie chart distribusi kelulusan
    print("\n--- Pie Chart Status Kelulusan ---")
    
    status_counts = df_hasil['Status'].value_counts()
    
    plt.figure(figsize=(7, 7))
    plt.pie(
        status_counts, 
        labels=status_counts.index, 
        autopct='%1.1f%%', 
        startangle=90, 
        colors=['lightgreen', 'lightcoral']
    )
    plt.title('Persentase Status Kelulusan Mahasiswa')
    plt.axis('equal') # Memastikan pie chart berbentuk lingkaran
    plt.show()
```
