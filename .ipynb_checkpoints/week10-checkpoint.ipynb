## Enkapsulasi (Encapsulation)

```python
class AkunBank:
    def __init__(self,saldo):
        self.__saldo = saldo # atribut private
        self.nilai = 10
    def lihat_saldo(self):
        return self.__saldo

    def setor(self, jumlah):
        self.__saldo += jumlah
        
    def tarik(self, jumlah):
        self.__saldo -= jumlah
```

```python
bank = AkunBank(50)
print(bank.lihat_saldo())
setor_lagi = bank.setor(80)
print(bank.lihat_saldo())
bank.tarik(80)
print(bank.lihat_saldo())
```

## Abstraksi (Abstraction)

```python
# class abstrak
class Bentuk(ABC):
    @abstractmethod
    def luas(self):
        pass

# class turunan konkret
class Persegi(Bentuk):
        def __init__(self, sisi):
            self.sisi = sisi
        
        def luas(self):
            return self.sisi * self.sisi

# --- Program Utama ---
# # Membuat objek persegi (karena bentuk tidak bisa dibuat objek langsung)
p = Persegi(5)
print("Luas Persegi: ", p.luas())
```

## Inheritance (Pewarisan)

```python
class Manusia:
    def __init__(self, nama): #method
        self.nama = nama #class

class Mahasiswa(Manusia):
    def __init__(self, nama, nim):
        super().__init__(nama)
        self.nim = nim

    def nama_nim(self):
        return self.nama, self.nim

manusia2 = Manusia("Agung")
mahasiswa2 = Mahasiswa("lala", "122400080")
print(mahasiswa2.nama_nim())
```

## Polimorfisme (Polyphism)

```python
class Hewan:
    def suara(self):
        pass 

class Kucing(Hewan):
    def suara(self):
        return "Meong"

class Anjing(Hewan):
    def suara(self):
        return "Guk guk"

kci = Kucing()
print(kci.suara())

anj = Anjing()
print(anj.suara())
```

## Percabangan (Nested If Statements)

```python
x = 41

if x > 10:
  print("Above ten,")
  if x > 20:
    print("and also above 20!")
  else:
    print("but not above 20.")
```

```python
age = 25
has_license = True

if age >= 18:
  if has_license:
    print("You can drive")
  else:
    print("You need a license")
else:
  print("You are too young to drive")
```

```python
score = 85
attendance = 90
submitted = True

if score >= 60:
  if attendance >= 80:
    if submitted:
      print("Pass with good standing")
    else:
      print("Pass but missing assignment")
  else:
    print("Pass but low attendance")
else:
  print("Fail")
```

```python
username = "Emil"
password = "python123"
is_active = True

if username:
  if password:
    if is_active:
      print("Login successful")
    else:
      print("Account is not active")
  else:
    print("Password required")
else:
  print("Username required")
```

```python
score = 92
extra_credit = 5

if score >= 90:
  if extra_credit > 0:
    print("A+ grade")
  else:
    print("A grade")
elif score >= 80:
  print("B grade")
else:
  print("C grade or below")
```
